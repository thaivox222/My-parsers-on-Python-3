```python
#нужно установить tqdm через pip [pip install -e git+https://github.com/tqdm/tqdm.git@master#egg=tqdm]

import requests
from  bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
#from selenium.webdriver import Chrome
#Chrome(executable_path ='C:/Users/Vladimir/Webdriver/bin/chromedriver.exe')
#driver = Chrome()
driver = webdriver.Chrome(executable_path ='C:/Users/Vladimir/Webdriver/bin/chromedriver.exe')
```

```python
params_list = ['brand', 'model', 'desc', 'base', 'shape', 'price',
              'p', 'lm_prc', 'lm', 'eff_yellow', 'eq', 'color', 'cri yellow', 'angle', 'flicker green', 'rating yellow']
```

```python
df = pd.DataFrame(columns = params_list)
from tqdm import tqdm_notebook #это прогресс бар. его можно и не использовать, так для красоты только
pagination = []
x=0
for page in tqdm_notebook(range(0, 1000,50), desc = 'Page loop'):# формируем урд пагинации в цикле. desc - это название прогресс бара
    #строка выше - (range(0, 3000,50) это перебор от 0 до 3000 с шагом 50
    temp_df = pd.DataFrame(columns = params_list) #создается дф, где колонки - список из params_list
    url = f'http://lamptest.ru/search/#currency=rub&type=LED&start={page}'#здесь получаем готовый урл
    driver.get(url) #открываем веб-страницу (driver.get)
    html = driver.page_source #получаем ее содержимое (driver.page_source)
    soup = bs(html, 'lxml') #передаем содержимое страницы в парсер BeautifullSoup/lxml
    x = x+50
    for cl in tqdm_notebook(params_list, desc='Params loop', leave=True): #desc - это название прогресс бара, leave - это новая итерация в том же баре или в новом
        res = soup.find_all("td", class_=cl)
        data = {idx:p.text for p, idx in zip(res, list(range(len(res))))}#итератор, куда складываются значения таблицы в виде словаря
        #data - это словарь словарей, в котором каждый элемент словаря - это набор всехзначений из столбца.
        #т.е у него д.б. 16 ключей по кол-ву столбцов
        temp_df[cl] = pd.Series(data) #заполняем дф по столбцам = Серия состоящая из словаря (дата)
    temp_df = temp_df.reset_index(drop=True)# если не сделать резет индекса, то не будет работать конкат
    df = pd.concat([df, temp_df],verify_integrity=False)
    if df.shape[0] != x:
        break #добвил это условие, чтобы цикл вырубался, если последняя страница не добавилась.
df.reset_index(inplace=True, drop=True) #пересчитаем индекс заново
```

```python
'''В первом цикле мы пребираем в итераторе все страницы пагианции
и передаем каждую такую страницу в суп, где она парсится.
Из-за того, что страницы динамичские, использутеся вебдрайвер, который эмулирует работу браузера.
Суть вобщем та же - заставить сервер отдать тебе страницу и передать ее в суп.
После этого она парсится и дальше ее уже можно потрошить.'''

'''Во втором цикле уже парсятся нужные теги и их содержимое'''
```

```python
'''решение по добавлению промежуточного дф к итоговому.
Сначала использовал аппенд - не получалось, затем решил использовать конкат - то же самое.
Решением оказалось, что надо сбрасывать индекс в промежуточном дф перед добавлением.
Такой ход я видел в одном из уроков ittensive - там автор сбрасывал индекс, аппендил. а потом опять сбрасывал.
Я его тогда спросил - нахрена второй раз сбрасываешь - а он такой - ну да, можно не сбрасывать.
А это не сброс был, а просто пересчет.
Вобщем идея такая - если в двух дф идет одинаковая нумерация по индексу - его надо сбрасывать иначе ни ошибки не выдает,
ни саму операцию не производит. Сиди и думай, что за  хрень такая. А это просто 2 индекса конфликтуют.
Поэтому перед склейкой нужно его сбрасывать.
Как только сделал - сразу получилось.
Оставил на конкате т.к. он у меня был последним при тестировании.
Скорее всего аппенд также будет работать, но мне лень проверять,т.к. много времени потратил'''
```
